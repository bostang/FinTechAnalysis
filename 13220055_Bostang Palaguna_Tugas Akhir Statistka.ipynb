#### perintah 3 : Tentukan confidence interval 95% untuk proporsi aktivitas online yang meningkat berikut:

### Selang kepercayaan untuk estimasi proporsi : sampel tunggal

$$ \hat{p} - error < p < \hat{p} + error $$

dengan $ error = z_{\alpha/2}\sqrt{\frac{\hat{p}\hat{q}}{n}} $

##### 3.1. secara manual satu per satu

##### a. streaming video/film/tv

```python
p_hat = aktivitas['proporsi'][1]
q_hat = 1 - p_hat
alpha = 0.05

from scipy.stats import norm
z_alpha_per_2 = norm.ppf(1-alpha/2)
n = df.count()['responden'] # jumlah responden

from math import sqrt
error = z_alpha_per_2 * sqrt(p_hat*q_hat/n)
lower_bound = p_hat - error
upper_bound = p_hat + error

print(f"{lower_bound} < p < {upper_bound} ")
```

##### b. mengikuti kelas online


```python
p_hat = aktivitas['proporsi'][2]
q_hat = 1 - p_hat
alpha = 0.05

from scipy.stats import norm
z_alpha_per_2 = norm.ppf(1-alpha/2)
n = df.count()['responden'] # jumlah responden

from math import sqrt
error = z_alpha_per_2 * sqrt(p_hat*q_hat/n)
lower_bound = p_hat - error
upper_bound = p_hat + error

print(f"{lower_bound} < p < {upper_bound} ")
```

##### c. konsultasi kesehatan online

```python
p_hat = aktivitas['proporsi'][4]
q_hat = 1 - p_hat
alpha = 0.05

from scipy.stats import norm
z_alpha_per_2 = norm.ppf(1-alpha/2)
n = df.count()['responden'] # jumlah responden

from math import sqrt
error = z_alpha_per_2 * sqrt(p_hat*q_hat/n)
lower_bound = p_hat - error
upper_bound = p_hat + error

print(f"{lower_bound} < p < {upper_bound} ")
```

##### 3.2. langsung di dataframe

```python
alpha = 0.05
aktivitas['z(alpha/2)'] = norm.ppf(1-alpha/2)
aktivitas['error'] = aktivitas['z(alpha/2)']*(aktivitas['proporsi']*(1-aktivitas['proporsi'])/n)**(1/2)
aktivitas['batas bawah'] = aktivitas['proporsi'] - aktivitas['error']
aktivitas['batas atas'] = aktivitas['proporsi'] + aktivitas['error']

# streaming video
print(f"streaming video:\n{aktivitas['batas bawah'][1]} < p < {aktivitas['batas atas'][1]}\n")

# kelas online
print(f"kelas online:\n{aktivitas['batas bawah'][2]} < p < {aktivitas['batas atas'][2]}\n")

# konsultasi kesehatan online
print(f"konsultasi kesehatan online:\n{aktivitas['batas bawah'][4]} < p < {aktivitas['batas atas'][4]}\n")
```

#### perintah 5 : Tentukan confidence interval 95% selisih proporsi:

### Selang kepercayaan estimasi proporsi sampel ganda (untuk jumlah sampel besar)

$$(\hat{p}_1-\hat{p}_2)-error < p_1-p_2 <(\hat{p}_1-\hat{p}_2)+error$$
dengan $error = z_{\alpha/2}\sqrt{\frac{\hat{p}_1 \hat{q}_1}{n_1}+\frac{\hat{p}_2 \hat{q}_2}{n_2}}$

a. pengguna kredivo - pengguna cicil

```python
alpha = 0.05
phat_kredivo = pilihan_fintech_lender['proporsi'][0]
qhat_kredivo = 1-phat_kredivo
phat_cicil = pilihan_fintech_lender['proporsi'][2]
qhat_cicil = 1- phat_cicil
n_kredivo = fintech['frekuensi'][1]
n_cicil = n_kredivo

kredivo_cicil = pilihan_fintech_lender.at[0,'proporsi'] - pilihan_fintech_lender.at[2,'proporsi'] # phat_kredivo - phat_cicil
error = norm.ppf(1-alpha/2) * (phat_kredivo * qhat_kredivo / n_kredivo + phat_cicil * qhat_cicil / n_cicil)**(0.5)

batas_bawah = kredivo_cicil - error
batas_atas = kredivo_cicil + error
print('{} < p_kredivo- p_cicil < {}'.format(batas_bawah, batas_atas))
```

##### b. meminjam untuk modal usaha -meminjam untuk belanja online

```python
alpha = 0.05
phat_modalUsaha = tujuan_fintech_lending['proporsi'][0]
qhat_modalUsaha = 1-phat_modalUsaha
phat_belanjaOnline = tujuan_fintech_lending['proporsi'][2]
qhat_belanjaOnline = 1- phat_belanjaOnline
n_modalUsaha = fintech['frekuensi'][1]
n_belanjaOnline = n_modalUsaha

modalUsaha_belanjaOnline = tujuan_fintech_lending.at[0,'proporsi'] - tujuan_fintech_lending.at[2,'proporsi'] # phat_modalUsaha - phat_belanjaOnline
error = norm.ppf(1-alpha/2) * (phat_modalUsaha * qhat_modalUsaha / n_modalUsaha + phat_belanjaOnline * qhat_belanjaOnline / n_belanjaOnline)**(0.5)

batas_bawah = modalUsaha_belanjaOnline - error
batas_atas = modalUsaha_belanjaOnline + error
print('{} < p_modalUsaha- p_belanjaOnline < {}'.format(batas_bawah, batas_atas))
```

#### 4.c. proporsi tujuan penggunaan layanan fintech lending

Dari 123 responden yang mengaku menggunakan fintech lending, 2 tujuan mendominasi. Teradpat 34 orang menggunakannya untuk membeli keperluan sehari-hari, 39 orang menggunakannya untuk berbelanja di marketplace(online shop). Sedangkan 3 tujuan lain yang ditanyakan hanya digunakan oleh sebagian kecil responden pengguna fintech : 22 orang menggunakan fintech lending untuk modal usaha, 14 orang menggunakan fintech lending untuk pembiayaan pendidikan, dan 9 orang menggunakannya untuk membeli kendaraan.

Kalau kita totalkan penggunaan fintech lending untuk masing-masing tujuan, kita hanya memperoleh : 22 + 34 + 39 + 14 + 9 = 118 < 123. Ini artinya ada responden yang saat mengisi survey mengaku menggunakan/tidak sengaja memilih opsi ‘menggunakan fintech lending’ padahal mereka sebenarnya tidak menggunakannya.

Fakta ini mengimplikasikan bahwa hasil analisis untuk survei fintech lending kita kali ini tidak akan kuat.

#### 4.d. Proporsi nominal peminjaman dalam jasa fintech lending

Dari 123 responden yang mengaku menggunakan jasa fintech lending, 54 orang (43 persen) menggunakannya dengan nominal kurang dari 1 juta dan 31 orang menggunakannya dengan nominal di antara 1 sampai dengan 5 juta. Nominal ini cukup masuk akal untuk penggunaan seperti belanja keperluan sehari-hari dan juga perbelanjaan di marketplace(mayoritas pada hasil 4.c).

Saya menduga orang-orang yang kebingungan saat mengisi survey (tidak sengaja memilih opsi menggunakan fintech lending padahal tidak menggunakannya)  berpikir bahwa opsi penggunaan ‘<1 juta’ memiliki arti yang sama dengan ‘tidak menggunakan fintech lending’. Sehingga opsi nominal ‘<1 juta’ mendapat bobot yang berlebih daripada seharusnya.

#### 4.e. Keluhan pengguna fintech lending

Dari 123 orang responden yang mengaku menggunakan jasa fintech lending, paling banyak terdapat 22 + 24 + 8 = 54 orang yang memiliki keluhan dalam menggunakan jasa fintech lending sedangkan sisanya tidak memiliki keluhan / tidak terpikirkan apa keluhannya. Sekitar 20 keluhan yang ada terkait dengan dengan aplikasi yang sulit diakses dan juga bunga tinggi, dan 8 keluhan terkait dengan ketidakmampuan untuk mengembalikan uang yang dipinjam ( terror oleh debt collector).
